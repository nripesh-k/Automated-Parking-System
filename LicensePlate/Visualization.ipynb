```python
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.datasets import cifar10
from keras.layers import Dense, Dropout, Activation, Conv2D, MaxPooling2D, Flatten
from keras.utils import np_utils
from keras.preprocessing import sequence
from keras import backend as K
from keras.utils.vis_utils import model_to_dot, plot_model
import model as mdl
import prepareData as pd
```

```python
def showData(data, row, col, n):
    plt.subplot(row, col, n)
    plt.imshow(data, cmap=plt.cm.gray_r)
    plt.axis('off')

def showLayer(data, row, col):
    col = 10
    for j in range(row):
        plt.figure(figsize=(16,16))
        for i in range(col):
            plt.subplot(row, col, i+1)
            plt.imshow(data[j,:,:,i],cmap=plt.cm.gray_r)
            plt.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
data,_ = pd.readHDF5('test.h5')
np.random.shuffle(data)
```

```python
recognitionModel = mdl.model()
recognitionModel.load_weights("trainedModel.h5")
```

```python
# original input images
show_size = 10
plt.figure(figsize=(20,20))
for i in range(show_size):
    showData(data[i], 1, show_size, i+1)
plt.show()
```

```python
get_first_layer_output = K.function([recognitionModel.layers[0].input], [recognitionModel.
layers[1].output])
first_layer = get_first_layer_output([data[0:show_size]])[0]
print(first_layer.shape)
showLayer(first_layer, first_layer.shape[0], first_layer.shape[3])
```
